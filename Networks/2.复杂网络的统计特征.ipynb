```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 常见的度分布：柏松分布与幂律分布
# 柏松分布：ER随机网络为例

```python
# 生成ER随机网络
n = 10000
p = 0.001
ER = nx.erdos_renyi_graph(n,p)
```

```python
# 获取平均度
d = dict(nx.degree(ER))
print('平均度为：',sum(d.values())/len(ER.nodes))
```

```python
# 获取所有可能的度值对应度概率
x = list(range(max(d.values())+1))
y = [i/n for i in nx.degree_histogram(ER)]
```

```python
# 绘制度分布
plt.plot(x,y,'ro-')
plt.xlabel('$k$')
plt.ylabel('$p_k$')
```

# 幂律分布：以BA无标度网络为例

```python
# 生成BA随机网络
n = 10000
m = 3
BA = nx.barabasi_albert_graph(n, 3)
```

```python
# 获取平均度
d = dict(nx.degree(BA))
print('平均度为：', sum(d.values()) / len(BA.nodes))
```

```python
# 获取所有可能的度值对应度概率
x = list(range(max(d.values()) + 1))
y = [i / n for i in nx.degree_histogram(BA)]
```

```python
# 绘制度分布
plt.plot(x, y, 'ro-')
plt.xlabel('$k$')
plt.ylabel('$p_k$')
```

```python
# 对数坐标轴
plt.plot(x, y, 'ro-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k$')
plt.ylabel('$p_k$')
```

```python
# 排除0（log后负无穷）
new_x = []
new_y = []
for i in range(len(x)):
    if y[i] != 0:
        new_x.append(x[i])
        new_y.append(y[i])
plt.plot(new_x, new_y, 'ro-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k$')
plt.ylabel('$p_k$')
```
