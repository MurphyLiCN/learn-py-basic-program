```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
```

```python
a = np.arange(0.0,5.0,0.02)
plt.subplot(2,2,1)
plt.plot(a,f(a))

plt.subplot(2,2,2)
plt.plot(a,np.cos(2*np.pi*a),'r--')

plt.subplot2grid((2,2),(1,0),colspan=2)
plt.plot([0,2,4,6,8],[3,1,4,5,2])
```

```python
a = np.arange(0.0,5.0,0.02)

gs = gridspec.GridSpec(2,2)
plt.subplot(gs[0,0])
plt.plot(a,f(a))

plt.subplot(gs[0,1])
plt.plot(a,np.cos(2*np.pi*a),'r--')

plt.subplot(gs[1,:])
plt.plot([0,2,4,6,8],[3,1,4,5,2])
```
